```
import pandas as pd
import gzip
import nltk
import gensim
from gensim.models import Word2Vec
from nltk.corpus import stopwords
from itertools import combinations
import string
from string import digits
from nltk import word_tokenize
```

```
data = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz" , lines=True)
```

```
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
stop_words
```

```
def unique_words(text):
  text = text.translate(text.maketrans('','',string.punctuation))
  text = text.translate(text.maketrans('','',digits))
  tokens = word_tokenize(text)
  result = [word for word in tokens if word not in stop_words]
  return result
```

```
nltk.download('punkt')
data['reviewText_unique_words']= data['reviewText'].apply(unique_words)
```

```
data['reviewText_unique_words'][1]
```

```
list = [[word] for word in data['reviewText_unique_words'][0]]
list
```

```
model=gensim.models.Word2Vec(list, window=5, min_count=1, workers=4)
```

```
for i in list:
  for j in list:
    if j>i:
      print(i[0],j[0],model.similarity(i[0],j[0]))
    else:
      continue
```
